```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
```

```python
def regress(data, yvar, xvars):
    Y=data[yvar]
    X=data[xvars]
    X['intercept']=1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
close_px=pd.read_csv('stock_px_2.csv', parse_dates=True, index_col=0)
```

```python
rets=close_px.pct_change().dropna()
```

```python
rets
```

```python
get_year=lambda x: x.year
```

```python
by_year=rets.groupby(get_year)
```

```python
by_year['AAPL'].describe()
```

```python
by_year[['SPX']].describe()
```

```python
#if rets=rets=close_px.pct_change() but not close_px.pct_change().dropna(),
#will trigger an error here due to NaN value...
by_year.apply(regress, 'AAPL', ['SPX'])
```
