```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=pd.read_csv('tips.csv')
```

```python
party_counts=pd.crosstab(tips['day'], tips['size'])
```

```python
party_counts
```

```python
party_counts2=party_counts.loc[:,2:5]
```

```python
party_counts2
```

```python
party_pcts=party_counts2.div(party_counts2.sum(1), axis=0)
```

```python
party_pcts
```

```python
party_pcts.plot.bar()
```
