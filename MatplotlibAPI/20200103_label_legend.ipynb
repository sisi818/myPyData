```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure()
```

```python
ax=fig.add_subplot(1,1,1)
```

```python
ax.plot(np.random.randn(1000).cumsum(),'ko', label='one')
```

```python
ax.plot(np.random.randn(1000).cumsum(),'r--', label='two')
```

```python
ax.plot(np.random.randn(1000).cumsum(),'b.', label='three')
```

```python
ax.legend(loc='best')
```

```python
ax.text(200,35,'Hello My first Plot!',
       family='monospace', fontsize=11)
```
