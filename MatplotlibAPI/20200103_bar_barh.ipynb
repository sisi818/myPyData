```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig, axes=plt.subplots(2,1)
```

```python
data=pd.Series(np.random.rand(16),index=list('abcdefghijklmnop'))
```

```python
data.plot.bar(ax=axes[0], color='k', alpha=0.7)
```

```python
data.plot.barh(ax=axes[1], color='k', alpha=0.5)
```
