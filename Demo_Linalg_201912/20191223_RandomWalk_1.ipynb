```python
import random
import matplotlib.pyplot as plt
```

```python
position=0
```

```python
walk=[position]
```

```python
steps=1000
```

```python
for i in range(steps):
    step=1 if random.randint(0,1) else -1
    position+=step
    walk.append(position)
```

```python
plt.plot(walk[:10000])
```
