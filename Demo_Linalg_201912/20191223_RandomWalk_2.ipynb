```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
nsteps=1000
```

```python
draws=np.random.randint(0,2,size=nsteps)
```

```python
steps = np.where(draws>0,1,-1)
```

```python
walk=steps.cumsum()
```

```python
plt.plot(walk[:1000])
```

```python
walk.min()
```

```python
walk.max()
```

```python
(np.abs(walk)>=10).argmax()
```
